```python
import pandas as pd
import os
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# episodes = pd.DataFrame()
episodes = []
episode_lengths = []
num_episodes = 71
```

```python
for ep in range(1,num_episodes):
    path = "data/"+str(ep)+".md"
    # Read data into data
    with open (path, "r") as f:
        data = f.readlines()
        episode_lengths.append(len(data[0]))
    episodes.append(data)
```

From the lengths, I can see that podcasts 44, 65 and 72 are missing transcripts, since their letter count is around an order of magnitude less than the others (3097 vs 40000 - 90000), but each podcast is around 1 hour long in length. 

Let's graph that to see the lengths.

## Lengths of the Episodes

```python
arr = list(range(1,73))
plt.bar(arr, episode_lengths)
```
